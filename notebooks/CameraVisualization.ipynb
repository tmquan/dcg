```python
from typing import Sequence, Tuple
import torch 
from pytorch3d.vis.plotly_vis import plot_scene, AxisArgs
from pytorch3d.renderer.cameras import (
    CamerasBase, 
    FoVPerspectiveCameras,
    look_at_view_transform
)
from pytorch3d.renderer import (
    RayBundle,
)

# Construct the surrounding cameras
cameras =[]

for k in range(0, 360, 45):
    dist = 4.0 * torch.ones(1)
    elev = torch.zeros(1) * 180 - k
    azim = torch.zeros(1) * 360
    R, T = look_at_view_transform(dist=dist, elev=elev, azim=azim)
    camera = FoVPerspectiveCameras(
        R=R, 
        T=T, 
        fov=45, 
        aspect_ratio=1
    )
    cameras.append(camera)
```

```python
# visualization of the cameras
plot = plot_scene(
    {"k": {i: camera for i, camera in enumerate(cameras)}}, 
    camera_scale=1,
    axis_args=AxisArgs(
        showgrid=False,
        zeroline=True,
        showline=True,
        showticklabels=True,
        showaxeslabels=True,
        backgroundcolor="rgb(200,230,200)",
    )  # kwarg axis_args
)
plot.layout.scene.aspectmode = "data"
plot
```
